```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


visa_df= pd.read_csv('Visadataset.csv')
cat_cols=visa_df.select_dtypes(include='object').columns
num_cols=visa_df.select_dtypes(exclude='object').columns

cert_cnt,den_cnt,index=[],[],[]
unique_continent=visa_df['continent'].unique()
columns=visa_df['case_status'].unique()
unique_continent
for i in unique_continent:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified'
    con3=visa_df['case_status']=='Denied'
    con4=con1&con2
    con5=con1&con3
    cert_cnt.append(len(visa_df[con4]))
    den_cnt.append(len(visa_df[con5]))
    
cols=visa_df['case_status'].unique()
pd.DataFrame(zip(cert_cnt,den_cnt), columns=cols,index=unique_continent)

    
#con1=visa_df['continent']=='Asia'
#con2=visa_df['case_status']=='Certified'
#con3=con1&con2
#visa_df[con3]
```
