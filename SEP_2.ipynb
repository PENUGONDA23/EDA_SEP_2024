```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

visa_df=pd.read_csv('Visadataset.csv')
cat_cols=visa_df.select_dtypes(include='object').columns
num_cols=visa_df.select_dtypes(exclude='object').columns
```

```python

visa_df=pd.read_csv('Visadataset.csv')
data=visa_df['continent'].value_counts()
keys=data.keys()
values=data.values
cols=['Labels','count']
df=pd.DataFrame(zip(keys,values),columns=cols)
plt.figure(figsize=(10,5))
plt.xlabel('Lables')
plt.ylabel('Count')
plt.title('Bar chart')
plt.bar(df['Labels'],df['count'])
plt.savefig('Continent_bat_chart.png')
plt.savefig('Continent_bat_chart.jpeg')
plt.show()
```

**Bar chart**

- For categorical columns we need to represent 2 ways

- Frequency table (tabular representation)

- bar chart (graphical representation)

- we know that bbar charts have 2 axis

    - X-axis: represents labels

    - Y-axis: represents count

- so in order to plot a bar chart first we need to make a frequency table

- here we will create the frequency table and immediately will draw the bar charts

- baar plot is available in **matplotlib package**





**Note:plt.show always at last**

- step 1: we read the data

- step 2: we take one cat column

- step 3: we created a frequency table

- step 4: we plot the bar chart

**count plot**

- count plot also same as bar chart

- count plot under seaborn package

- count plot does not require frequency table to plt bar chart

- count plot need only column name annd original dataframe name

- Assume that column name:**continent**

- Original data frame name **visa_df**

```python
# count plot by default give alphebatical order
visa_df=pd.read_csv('Visadataset.csv')
sns.countplot(data=visa_df,x='continent')
plt.title('countplot')
plt.show()
```

```python
# more parameters for same seaborn package
order=['Asia','Europe','North America','South America','Africa','Ocenia']
keys=visa_df['continent'].value_counts().keys()
plt.figure(figsize=(10,5))
visa_df=pd.read_csv('Visadataset.csv')
sns.countplot(data=visa_df,x='continent',order=keys)
plt.title('countplot')
plt.savefig('Continent_count_plot_usingsns.jpeg')
plt.show()
```
