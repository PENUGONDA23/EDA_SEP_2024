```python
# step 1 read continent data
#step-2 we want asia so the data equal to Asia
#5==5 o/p True
visadf['continent']
con=visadf['continent']== 'Asia'
# we are passing 'visadf['continent']== 'Asia' as condition to visadf object which get all rows with 'Asia' continent
visadf[con]
# to get lenngth that is rows of dataframe len(visadf(con))
```

```python
unique_labels=visadf['continent'].unique()
for i in unique_labels:
    visadf['continent']
    condition=visadf['continent']==i
    visadf[condition]
    count=len(visadf[condition])
    print(f'the number of applicants from {i} is {count}')
```

```python
#create dataframe out of continents & their count
unique_labels=visadf['continent'].unique()
count=[]
for i in unique_labels:
    visadf['continent']
    condition=visadf['continent']==i
    visadf[condition]
    count.append(len(visadf[condition]))
pd.DataFrame(zip(count,unique_labels),columns=['continent','number of applicants'])
```
