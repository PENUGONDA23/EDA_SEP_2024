**Label Encoder**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

visa_df=pd.read_csv('Visadataset.csv')
cat_cols=visa_df.select_dtypes(include='object').columns
le=LabelEncoder()
for i in cat_cols:
    visa_df[i]=le.fit_transform(visa_df[i])
visa_df


```

```python
visa_df=pd.read_csv('Visadataset.csv')
from sklearn.preprocessing import LabelEncoder
le.LabelEncoder()
data=le.fit_transform(visa_df['education_of_employee'])
le.inverse_transform(data)

```
